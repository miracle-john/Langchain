# 대규모 언어 모델(Large Language Model, LLM)의 문제점
- 일반적인 상황에서 프롬프트에 응답하는 데 탁월하지만, <font color=red>훈련한 적이 없는 특정 영역에서 성능이 떨어짐</font>
- 상태를 저장하지 않으므로 이전 <font color=red>대화 내용을 기억하지 못함</font> -> 데이터베이스의 필요
- 일률적인 규칙이 없으므로 감정 분석, 분류, 질문 답변과 요약 등 <font color=red>서로 다른 시나리오에 특화된 다양한 모델을 사용</font>해야 함
- 이를 위해서 기계 학습 엔지니어가 LLM을 조직의 내부 데이터 소스와 통합하고 프롬프트 엔지니어링을 적용해야 함
- LangChain은 이러한 데이터 응답 애플리케이션을 개발하기 위한 중간 단계를 간소화하여 프롬프트 엔지니어링의 효율성을 높임
  - LLM 기반 어플리케이션을 구축하기 위해서는 프롬프트를 보내고 응답을 기다리는 것 등의 작업이 필요 (매개변수 조정, 프롬프트 보강, 응답 조정 등)


# LangChain 개요

- 2022년 10월 말 해리슨 체이스가 제작
- <font color=red>LLM과 애플리케이션의 통합</font>을 간소화하도록 설계된 강력한 프레임워크
- 간단하고 통합된 API를 활용하여 개발자들이 코드를 크게 변경하지 않고 <font color=red>모델을 쉽게 교체하거나 대체</font>할 수 있도록 함
- 검색 증강 생성(RAG) 어플리케이션 제작, 구조화된 데이터 분석, 챗봇 등을 제작 가능
- 주요 기능
  - <font color=red>문맥 인식</font>: 언어 모델과 다양한 문맥 소스(프롬프트 지시, 예제, 응답의 근거 내용 등)를 연동하며, 사용자의 문맥을 정확히 이해
  - <font color=red>추론 능력</font>: 제공된 문맥에 기반하여 어떤 대답을 할지, 또는 어떠한 액션을 취할지에 대한 추론이 가능

  - <font color=red>통합된 import 구조</font> : 라이브러리를 기능별로 정리
  - <font color=red>TypedDict 기반 상태 관리</font> : 에이전트 상태를 TypedDict로 정의해야 함 (Pydantic, dataclass 사용 불가)

# LangChain 1.0 프레임워크 구성
- **에이전트(Agent) 중심**의 프레임워크로 재설계
- 복잡한 체인 구성 대신, create_agent를 통해 10줄 이내의 코드로 프로덕션 수준의 AI 에이전트 구축 지원

- <font color=red><b>langchain</b></font> (메인 패키지) : 에이전트 생성, 모델 초기화, 도구 정의 등 핵심 기능을 제공

  <pre>
  from langchain.agents import create_agent          # 에이전트 생성
  from langchain.chat_models import init_chat_model  # 통합 모델 초기화
  from langchain.tools import tool                   # 도구 데코레이터
  </pre>

- <font color=red>langchain-core</font> : 메시지, 프롬프트, Runnable 등 기본 추상화와 인터페이스를 제공

  <pre>
  from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
  from langchain_core.prompts import ChatPromptTemplate
  </pre>

- <font color=red>프로바이더 패키지</font> (langchain-openai, langchain-anthropic 등) : 각 LLM 제공자와의 통합을 담당  

  <pre>
  from langchain_openai import ChatOpenAI
  from langchain_anthropic import ChatAnthropic
  from langchain_google_genai import ChatGoogleGenerativeAI
  </pre>

- <font color=red>LangGraph</font> :  
  - 에이전트의 실행 런타임을 제공
  - create_agent로 생성된 에이전트는 내부적으로 LangGraph 위에서 동작

- <font color=red>LangSmith</font> : 에이전트의 실행을 추적, 디버깅, 평가하는 관측성(Observability) 플랫폼    

# LangChain 아키텍처
- 에이전트 중심 아키텍처
- create_agent()의 주요 파라미터
  - **model** : 사용할 LLM 모델 (init_chat_model 또는 ChatOpenAI 등)
  - **tools** : 에이전트가 사용할 도구 리스트
  - **system_prompt** : 시스템 프롬프트 (에이전트의 역할 정의)
  - **middleware** : 실행 흐름을 제어하는 미들웨어 리스트
  - **checkpointer** : 상태 영속성을 위한 체크포인터 (메모리 기능)
  - **response_format** : 구조화된 출력 형식 (ToolStrategy, ProviderStrategy)

# LangChain 생태계 구성
- 개발도구
  - <font color=red>LangChain</font> : 에이전트 및 LLM 애플리케이션 개발 프레임워크
  - <font color=red>LangGraph</font> : 복잡한 워크플로우를 위한 저수준 오케스트레이션 프레임워크
  - <font color=red>LangSmith</font> : 디버깅, 테스트, 평가, 모니터링 플랫폼

- 배포도구
  - <font color=red>LangServe</font> : LangChain 애플리케이션을 REST API로 배포
  - <font color=red>LangGraph Platform</font> : 에이전트를 클라우드에 배포하고 관리

- 통합 패키지
  - <font color=red>langchain-openai</font> : OpenAI 모델 통합
  - <font color=red>langchain-anthropic</font> : Claude 모델 통합
  - <font color=red>langchain-google-genai</font> : Google Gemini 통합
  - <font color=red>langchain-community</font> : 커뮤니티 기여 통합

# LangChain의 장점
- **언어 모델 용도 변경**
  - 재훈련이나 미세 조정 없이 LLM을 분야별 애플리케이션에 맞게 재활용 가능
  - 개발팀은 독점 정보를 참조하여 모델 응답을 보강하는 복잡한 애플리케이션 구축 가능
  - RAG 워크플로를 만들 수 있으며 이를 통해 모델 할루시네이션을 줄이고 응답 정확도 향상
- **AI 개발 간소화**
  - LangChain은 복잡한 데이터 소스 통합과 프롬프트 세분화를 추상화하여 인공 지능(AI) 개발을 간소화
  - 개발자는 시퀀스를 사용자 지정하여 복잡한 애플리케이션 빠르게 구축 가능
- **개발자 지원**
  - LangChain은 언어 모델을 외부 데이터 소스와 연결하는 도구를 AI 개발자에게 제공
  - LangChain은 오픈 소스이며 활발한 커뮤니티 지원을 받음

## 기본 LLM 체인 (Prompt + LLM)
- LLM 기반 애플리케이션 개발에서 핵심적인 개념 중 하나
- **사용자의 입력(프롬프트)을 받아 LLM을 통해 적절한 응답이나 결과를 생성**하는 구조
- 대화형 AI, 자동 문서 생성, 데이터 분석 및 요약 등 다양한 용도로 활용

- 구성 요소
  - **프롬프트(Prompt)**
    - 사용자 또는 시스템에서 제공하는 입력으로, LLM에게 특정 작업을 수행하도록 요청하는 지시문
    - 질문, 명령, 문장 시작 부분 등 다양한 형태를 취할 수 있으며, LLM의 응답을 유도하는 데 중요한 역할

  - **LLM(Large Language Model)**
    - GPT, Gemini 등 대규모 언어 모델로, 대량의 텍스트 데이터에서 학습하여 언어를 이해하고 생성할 수 있는 인공지능 시스템
    - 프롬프트를 바탕으로 적절한 응답을 생성하거나, 주어진 작업을 수행하는 데 사용

- 일반적인 작동 방식
  - **프롬프트 생성**
    - 사용자의 요구 사항이나 특정 작업을 정의하는 프롬프트를 생성
    - 프롬프트는 LLM에게 전달되기 전에, 작업의 목적과 맥락을 명확히 전달하기 위해 최적화될 수 있음

  - **LLM으로 프롬프트 전달**
    - LLM은 제공된 프롬프트를 분석하고, 학습된 지식을 바탕으로 적절한 응답을 생성
    - 이 과정에서 LLM은 내부적으로 다양한 언어 패턴과 내외부 지식을 활용하여, 요청된 작업을 수행하거나 정보를 제공
  - **LLM의 응답 반환**
    - LLM에 의해 생성된 응답은 최종 사용자에게 필요한 형태로 변환되어 제공
    - 응답은 직접적인 답변, 생성된 텍스트, 요약된 정보 등 다양한 형태가 가능

# 프롬프트 탬플릿 (PromptTemplate) 구성 요소
- PromptTemplate : 단일 문장 또는 입력변수 등의 <font color=red>간단한 명령을 입력</font>하여 단일 문장 또는 <font color=red>간단한 응답을 생성</font>하는 데 사용되는 프롬프트를 구성할 수 있는 문자열 템플릿
  - <font color=red>지시</font> : 언어 모델에게 어떤 작업을 수행하도록 요청하는 구체적인 지시
  - <font color=red>예시</font> : 요청된 작업을 수행하는 방법에 대한 하나 이상의 예시
  - <font color=red>맥락</font> : 특정 작업을 수행하기 위한 추가적인 맥락
  - <font color=red>질문</font> : 어떤 답변을 요구하는 구체적인 질문

- 예시: 제품 리뷰 요약
  - 지시: "아래 제공된 제품 리뷰를 요약해주세요."
  - 예시: "예를 들어, '이 제품은 매우 사용하기 편리하며 배터리 수명이 길다'라는 리뷰는 '사용 편리성과 긴 배터리 수명이 특징'으로 요약할 수 있습니다."
  - 맥락: "리뷰는 스마트워치에 대한 것이며, 사용자 경험에 초점을 맞추고 있습니다."
  - 질문: "이 리뷰를 바탕으로 스마트워치의 주요 장점을 두세 문장으로 요약해주세요."


## 프롬프트 템플릿의 활용
  - 파라미터
    - <font color=red>template</font>: 문자열 템플릿 및 변수 설정 (중괄호 {} 사용)
    - <font color=red>input_variables</font> : PromptTemplate에서 사용되는 변수(중괄호 안에 들어갈 변수)의 이름을 정의하는 리스트

 - <font color=red>from_template()</font> : 문자열 템플릿을 직접 설정하는 함수

```python
# temperature : 0-1 -> 1에 가까울 수록 자유도가 높음
# max_tokens : 사용할 수 있는 최대 토큰 (입력토큰(질문) + 응답토큰 (AI의 응답))
llm = init_chat_model(model="gpt-4o-mini",
                      temperature=0,
                      max_tokens=500)

# 변수가 포함된 프롬프트
template = "{country}의 수도는 어디야 ?"

# PromptTemplate : 템플릿 -> 프롬프트 변환
# invoke() 사용하기 전에 변수값을 할당해서 프롬프트롤 생성한 후에 모델에 전달
# 방법 1
prompt1 = PromptTemplate(template=template, input_variables=["country"])

# 방법 2
prompt2 = PromptTemplate.from_template(template)
```

## LangChain Expression Language (LCEL)
- https://blog.langchain.dev/langchain-expression-language/
- <font color=red>체인을 구성하고 스트리밍, 배치 및 비동기 지원</font>을 기본적으로 제공하는 선언적 방법
- 기본적으로 파이썬 또는 타입스크립트/자바스크립트를 사용한 체인 만들기의 고수준 대안으로 코드를 사용해 구성할 때 사용하는 모든 기존 랭체인 생성자를 그대로 사용해 체인을 만들 수 있음

## | (파이프라인)으로 체인 생성하기 (LCEL)
  - LangChain 0.1.17 이후 버전부터 사용
  - <font color=red>|</font> 기호는 unix 파이프 연산자와 유사하며, <font color=red>서로 다른 구성 요소를 연결하고 한 구성 요소의 출력을 다음 구성 요소의 입력으로 전달</font>

  - prompt | llm | OutputParser와 같이 체인이 연결된 경우
<center>  
<img src="https://arome1004.cafe24.com/images/deeplearning/lcel1.png" width=60%>   
</center>

- 표준 인터페이스 함수을 이용하여 딕셔너리 형태로 값들을 전달
   - <font color=red>invoke()</font> : 주어진 입력에 대해 체인을 호출하고, 결과를 반환
   - <font color=red>batch()</font> : 입력 리스트에 대해 체인을 호출하고, 각 입력에 대한 결과를 리스트로 반환
   - <font color=red>stream()</font> : 입력에 대해 체인을 호출하고, 결과의 조각들을 스트리밍

```python
chain = prompt2 | llm

# (1) 변수값을 템플리의 변수에 할당해서 프롬프트를 완성
# (2) 완성된 프롬프트를 LLM에 전달
# (3) LLM이 응답 생성해서 반환
answer = chain.invoke({"country":"중국"})
print(answer)
print(answer.content)
```

# ChatPromptTemplate
- <font color=red>대화형 상황에서 여러 메시지 입력을 기반으로 단일 메시지 응답을 생성</font>하는 데 사용
- 대화형 모델이나 챗봇 개발에 주로 사용
- 입력은 여러 메시지를 원소로 갖는 리스트로 구성되며, 각 메시지는 <font color=red>역할(role)</font>과 <font color=red>내용(content)</font>으로 구성

- Message 유형
  - <font color=red>SystemMessage</font>: 모델의 행동 방식을 정의 - 대화의 맥락, 역할, 제약사항 등을 설정
  - <font color=red>HumanMessage</font>: 사용자의 입력 - 텍스트뿐 아니라 이미지 등 멀티모달 콘텐츠를 포함
  - <font color=red>AIMessage</font>: 모델의 응답 - 모델 호출 결과로 반환되며, 다양한 메타데이터를 포함
  - <font color=red>ToolMessage</font>: 도구 실행 결과를 모델에 전달할 때 사용

- <font color=red>from_messages()</font> : 2개의 튜플 형태 메시지 리스트를 입력 받아, 각 메시지의 역할(type)과 내용(content)을 기반으로 프롬프트를 구성  
  - 사용자의 입력을 프롬프트에 동적으로 삽입하여, 최종적으로 대화형 상황을 반영한 메시지 리스트를 생성
  - 시스템은 기능을 설명하고, 사용자는 관련 질문을 입력

- <font color=red>from_messages()</font> : 메시지 사용하기

```python
# 입력 파라미터와 반환값에 대한 타입을 설정
def encoding_image(image_path:str) -> str :
  # 파일을 열어서 base64로 인코딩, 한글은 utf-8로 디코딩
  with open(image_path, "rb") as image_file:
    #
    return base64.b64encode(image_file.read()).decode("utf-8")

img_data = encoding_image("./data/sample2.png")

llm = init_chat_model("gpt-4o")

message = HumanMessage(
    content = [
        {"type" : "text", "text" : "도표를 분석해줘"},
        {"type" : "image_url",
         "image_url" : {"url" : f"data:image/png;base64,{img_data}"}}
    ]
)

answer = llm.invoke([message])
display(Markdown(answer.content))
```

# RunnablePassthrough
- <font color=red>RunnablePassthrough</font>
  - 체인에서 데이터 흐름을 제어할 때 유용
  - **여러 개의 체인이 있을때 1개 이상 체인에 동일값을 명시적으로 전달**  
  - 특히 **질문이나 다른 입력이 동적으로 변경**될 가능성이 있는 경우에 많이 사용
- <font color=red>RunnablePassthrough.assign(...)</font>
  - 입력을 받아 assign 함수에 전달된 인수를 추가
  - 입력 값으로 들어온 값의 key/value 쌍과 새롭게 할당된 key/value 쌍을 합침
- <font color=red>RunnableParallel()</font> : 여러 Runnable 인스턴스를 병렬로 실행
- <font color=red>RunnableLambda()</font> : 사용자 정의 함수를 맵핑, 함수를 체인으로 연결할 수 있게 만들어 줌

- RunnablePassthrough()는 runnable 객체이며, runnable 객체는 invoke() 메소드를 사용하여 별도 실행이 가능

- RunnableLambda()

```python
llm = init_chat_model("gpt-4o-mini")

def get_today(a) :
  # %b-%d : 월이름 (바이너리), 일 (숫자)
  return datetime.today().strftime("%b-%d")

prompt = PromptTemplate.from_template("""{today}가 생일인 한국의 유명인 {n} 명을 나열해줘
이름과 생년월일을 같이 표시해줘""")

# RunnableLambda() : 함수를 호출해서 결과를 받아오는 기능
# RunnablePassthrough() : 설정한 값을 가져오는 기
chain = (
    {"today" : RunnableLambda(get_today), "n" : RunnablePassthrough()}
    | prompt | llm
)

answer = chain.invoke({"n" : 3})

display(Markdown(answer.content))
```

- RunnableBranch()
  - 조건문과 RunnableLambda()를 사용한 라우팅을 더 활용

```python
# 한국어, 영어로 질문하는 경우 각각에 맞는 프롬프트를 선택하도록 하는 기능
llm = init_chat_model("gpt-4o-mini")

k_prompt = ChatPromptTemplate.from_template(
    "한국어로 된 다음과 같은 질문에 대해 한국어로 대답해줘 : {question}"
)

e_prompt = ChatPromptTemplate.from_template(
    "Answer the flowing english question in English : {question}"
)

d_prompt = ChatPromptTemplate.from_template(
    "Please Answer : {question}"
)

# 질문이 대한 언어 인식 함수
def detect_language(input_dict) :
  # 딕셔너리에서 question 키 값을 가져온다 (사용자 질문)
  question = input_dict.get("question", "")

  # "\uac00" ~ "\ud7a3" : 한글에 해당하는 코드의 범위 -> 한글이라면
  if any("\uac00" <= char <= "\ud7a3" for char in question) :
    return "korean"
  return "english"

# 3가지 경우의 실행 분기체인 설정 (한국어, 영어, 기타)
chain = RunnableBranch(
    (lambda x : detect_language(x) == "korean", k_prompt | llm),
    (lambda x : detect_language(x) == "english", e_prompt | llm),
    d_prompt | llm
)

answer = chain.invoke({"question" : "인공지능이란"})
display(Markdown(answer.content))

answer = chain.invoke({"question" : "AI is ?"})
display(Markdown(answer.content))
```

- with_structured_output
  - LangChain 1.0에서 권장하는 구조화 출력 방식
  - 모델의 네이티브 JSON 스키마 기능을 활용

  - LLM 모델에 설정 구조 (class)로 출력하는 기능을 붙여주는 기능

```python
# 구조화된 형식 정의
# Field : 변수에 대한 설명이나 조건 등을 설정
class MovieReview(BaseModel) :
  """영화 리뷰 정보"""
  title : str = Field(description="영화 제목")
  rating : int = Field(description="영화 평점 (1-10점)")
  pros : str = Field(description="영화 장점")
  cons : str = Field(description="영화 단점")
  summary : str = Field(description="한 줄 요약")
```

```python
# 기존 LLM에 구조화된 출력을 만드는 기능을 추가
strudcted_llm = llm.with_structured_output(MovieReview)

answer = strudcted_llm.invoke("군체 영화에 대해 설명해줘")

print(f"제목 : {answer.title}")
print(f"평점 : {answer.rating}")
print(f"장점 : {answer.pros}")
print(f"단점 : {answer.cons}")
print(f"요약 : {answer.summary}")
```

- 실습 : 코드 리뷰 기능 구현
  - 코드를 입력받아서 해당 코드에 대해 분석하는 기능
  - 구조화된 데이터
    - 코드 기능 요약 (summary)
    - 1-10점 품질 평가 (score)
    - 문제점 (issues)
    - 개선할 내용 (suggestions)

```python
class CodeReview(BaseModel):
    """코드 리뷰 결과"""
    summary: str = Field(description="코드 기능 요약")
    score: int = Field(description="1-10점 품질 점수")
    issues: list[str] = Field(description="발견된 문제점 목록")
    suggestions: list[str] = Field(description="개선 제안 목록")
    suggest_code: str = Field(description="개선된 코드")
```

```python
# 구조화된 출력 형식을 llm에 연결
llm = init_chat_model("gpt-4o-mini")
structured_llm = llm.with_structured_output(CodeReview)

result = structured_llm.invoke("""
다음 코드를 리뷰해주세요:

def calc(x):
    return x*2+1

calc(3)
""")

print(f"코드 요약: {result.summary}")
print(f"점수: {result.score}/10")
print(f"문제점: {result.issues}")
print(f"개선점: {result.suggestions}")
print(f"개선코드: {result.suggest_code}")
```
